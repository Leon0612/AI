```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import cm as cm

# 用于设置中文坐标轴标题，英文坐标轴标题无需加上这两行代码
mpl.rcParams["font.sans-serif"]=["SimHei"]
# 不使用默认的Unicode minus模式来处理坐标轴轴线的刻度标签是负数的情况
mpl.rcParams["axes.unicode_minus"]=False
```

### 折线图和散点图

```python
x = np.linspace(0.5, 3.5, 100)
y = np.sin(x)
y1 = np.random.randn(100)
```

```python
# figsize --> 指定宽和高，单位英寸 
# dpi --> 表示分辨率(每英寸多少个像素)，默认值是80 
# facecolor --> 设定图形的背景颜色
# edgecolor --> 设定图形的边框颜色
# frameon --> 是否显示边框，True表示显示，False表示不显示
plt.figure(figsize=(12, 6), dpi=200, facecolor=None, edgecolor=None, frameon=True)


# scatter(): 散点图
# x, y --> x轴和y轴对应的数据
# c -- > 散点的颜色，默认是蓝色
# marker --> 散点的格式
# alpha --> 透明度
plt.scatter(x, y1, c="black", marker=None, alpha=None, linewidths=None, label="scatter figure")


# plot(): 曲线图
# x, y --> 表示x轴和y轴的数据
# label --> 表示曲线的标签
# format_string --> 表示控制曲线的格式字串，包括设定颜色、样式、宽度等
plt.plot(x, y, ls="-", lw=2, label="plot figure")


# xlim, ylim --> 设定x轴、y轴刻度范围
# xmin, xmax, ymin, ymax --> 设定x轴、y轴的最小值、最大值
plt.xlim(0.0, 4.0)
plt.ylim(-3.0, 3.0)


# xlabel(), ylabel(): 设置x轴、y轴的坐标标签
# fontsize --> 标签的大小
# rotation --> 标签的角度
plt.xlabel("x_axis", fontsize=12, rotation=None)
plt.ylabel("y_axis", fontsize=12, rotation=88)


# grid(): 网格
# axis --> axis='x'表示仅绘制x轴的网格线，axis='both'表示绘制x、y轴的网格线
# linestyle --> 设置网格线的样式
# linewidth --> 设置网格线的宽度
plt.grid(ls=":", color="r")


# axhline(): 绘制平行于x轴的水平参考线
# y --> 水平参考线的出发点
plt.axhline(y=0.0, c="r", ls="--", lw=2)


# axvspan(): 绘制垂直于x轴的参考区域
plt.axvspan(xmin=1.0, xmax=2.0, facecolor="y", alpha=0.5)


# annotate(): 添加注释
# s --> 表示注释的内容，以字符串的方式输入
# xy --> 表示标注的位置，以xy=(num1, num2)表示，num1表示x轴的刻度，num2表示y轴的刻度
# xytext --> 表示文本的位置，以xytext=(num1, num2)表示，同上
# arrowprops --> 设置箭头的特征, arrowprops=dict(width箭头宽度, frac箭头头部所占比例, headwith箭头底部宽度, shrink箭头收缩程度)
plt.annotate("maximum", xy=(np.pi/2, 1.0), xytext=((np.pi/2)+0.15, 1.5), weight="bold", color="r",
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color="r"))

plt.annotate("spines", xy=(0.75, -3), xytext=(0.35, -2.25), weight="bold", color="b",
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color="b"))

plt.annotate("", xy=(0, -2.78), xytext=(0.4, -2.32), weight="bold", color="b",
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color="b"))

plt.annotate("", xy=(3.5, -2.98), xytext=(3.6, -2.70), weight="bold", color="b",
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color="b"))


# text(): 设置文字说明
plt.text(3.6, -2.70, "'|' is tickline", weight="bold", color="b")
plt.text(3.6, -2.95, "3.5 is ticklabel", weight="bold", color="b")


plt.title("structure of matplotlib")


# legend(): 显示图例，通过loc=？来控制图例的位置
# 0 最佳 1 右上 2 左上 3 左下 4 右下 5 右 6 中左 7 中右 8 中下 9 中上 10 中 空白表示自动
plt.legend(loc="upper right")


# dpi --> 分辨率
# format --> 格式: png, jpg, pdf等
# transparent --> 用于将图片背景设置为透明
# plt.savefig(fname, dpi=None, format=None, transparent=False)


# 设置坐标轴的位置
plt.gca().xaxis.set_ticks_position("bottom")
plt.gca().yaxis.set_ticks_position("left")


# 将顶部和右侧的边框线去掉
for spine in plt.gca().spines.keys():
    if spine == "top" or spine == "right":
        plt.gca().spines[spine].set_color("none")


plt.show() 
```

### 柱状图

```python
plt.figure(figsize=(11, 15), dpi=300)

plt.subplot(321)
x1 = [1, 2, 3, 4, 5, 6, 7, 8]
y1 = [3, 1, 4, 5, 8, 9, 7, 2]
# align --> 柱体对齐方式
# tick_label --> 刻度标签值
# alpha --> 透明度
plt.bar(x1, y1, width=0.8, align="center", color="b", alpha=0.8, tick_label=list('abcdefgh'))
plt.xlabel("箱子编号", fontsize=7)
plt.ylabel("箱子重量(kg)", fontsize=7)
plt.title("柱状图")


plt.subplot(322)
x2 = [1, 2, 3, 4, 5]
y21 = [6, 10, 4, 5, 1]
y22 = [2, 6, 3, 8, 5]
plt.bar(x2, y21, align="center", color="r", alpha=0.8, tick_label=["A", "B", "C", "D", "E"], label="classA")
# bottom --> 柱状图y轴的起使位置，此处是上一个柱状的y
plt.bar(x2, y22, align="center", color="b", alpha=0.8, bottom=y21, label="classB")
plt.xlabel("test difficulty", fontsize=7)
plt.ylabel("paper nums", fontsize=7)
plt.title("堆积柱状图")
plt.legend()


plt.subplot(323)
# hatch --> 填充线
plt.barh(x1, y1, align="center", color="c", alpha=0.8, tick_label=list('abcdefgh'), hatch="/")
plt.xlabel("箱子编号", fontsize=7)
plt.ylabel("箱子重量(kg)", fontsize=7)
plt.title("条形图")


plt.subplot(324)
x4 = [1, 2, 3, 4, 5]
y41 = [6, 10, 4, 5, 1]
y42 = [2, 6, 3, 8, 5]
plt.barh(x4, y41, align="center", color="r", alpha=0.8, tick_label=["A", "B", "C", "D", "E"], label="classA")
plt.barh(x4, y42, align="center", color="b", alpha=0.8, left=y41, label="classB")
plt.ylabel("test difficulty", fontsize=7)
plt.xlabel("paper nums", fontsize=7)
plt.title("堆积条形图")
plt.legend()


plt.subplot(325)
x5 = np.arange(5)
y51 = [6, 10, 4, 5, 1]
y52 = [2, 6, 3, 8, 5]
bar_width = 0.35
tick_label = ["A", "B", "C", "D", "E"]
plt.bar(x5, y51, bar_width, color="c", align="center", label="classA", alpha=0.7)
plt.bar(x5+bar_width, y52, bar_width, color="b", align="center", label="classB", alpha=0.7)
plt.xlabel("test difficulty", fontsize=7)
plt.ylabel("paper nums", fontsize=7)
plt.title("分块图")
plt.xticks(x5+bar_width/2, tick_label)
plt.legend()


plt.subplot(326)
boxWeight = np.random.randint(0, 10, 100)
bins = range(0, 11, 1)
# bins --> 统计的区间分布
plt.hist(boxWeight, bins=bins, color="orange", rwidth=1, alpha=0.8)
plt.xlabel("box weight(kg)", fontsize=7)
plt.ylabel("sale amount", fontsize=7)
plt.title("直方图")
plt.show()


# 自动调整子图，使得标题与标签不重合等
plt.tight_layout()
plt.show()
```

### 饼图与气泡图

```python
# subplot(): 子图
# nrows --> 子图的行数
# ncols --> 子图的列数
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(5, 2), dpi=200)

# 设置主标题
# fig.suptitle('xxxxxx')

axes = axes.flatten()

# 选取第一个子图
ax1 = axes[0] 
kinds = ["a", "b", "c", "d"]
colors = ["red", "orange", "blue", "gray"]
nums = [0.05, 0.45, 0.15, 0.35]
# labels --> 每一块的文字说明
# autopct --> 控制饼图内百分比设置
# startangle --> 起始绘制角度,默认图是从x轴正方向逆时针画起,如设定=90则从y轴正方向画起
ax1.pie(nums, labels=kinds, autopct="%3.1f%%", startangle=60, colors=colors, textprops={'fontsize':6})
ax1.set_title("饼图", fontsize=7)


ax2 = axes[1]
a = np.random.randn(100)
b = np.random.randn(100)
# s --> 点的大小
# c --> 点的颜色，可以是一系列数字，通过cmap映射到颜色
# cmap --> 仅在c是浮点数时使用，表示颜色方案
ax2.scatter(a, b, s=np.power(2*a+4*b, 2), c=np.random.rand(100), cmap= mpl.cm.RdYlBu, marker="o")
# 设置刻度字体的大小
ax2.tick_params(labelsize=6)
ax2.set_title("气泡图", fontsize=7)

plt.tight_layout()
plt.show()
```

### 极限图

```python
plt.figure(dpi=100)
barSlices = 12
theta = np.linspace(0.0, 2*np.pi, barSlices, endpoint=False)
r = 30 * np.random.rand(barSlices)
plt.polar(theta, r, color="chartreuse", linewidth=2, marker="*", mfc="b", ms=10)
plt.show()
```

棉棒图

```python
plt.figure(dpi=100)
x = np.linspace(0.5, 2*np.pi, 20)
y = np.random.randn(20)
plt.stem(x, y, linefmt="-.", markerfmt="o", basefmt="-")
plt.show()
```

### 误差棒图

```python
plt.figure(dpi=100)
x = np.linspace(0.1, 0.6, 6)
y = np.exp(x)
plt.errorbar(x, y, fmt="bo:", yerr=0.2, xerr=0.02)
plt.xlim(0, 0.7)
plt.show()
```

### 3D图形

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

figure = plt.figure(dpi=100)
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=mpl.cm.coolwarm, linewidth=0.1)
plt.show()
```

```python
plt.figure(dpi=100)
plt.contourf(X, Y, Z, cmap=mpl.cm.coolwarm)
plt.colorbar()
plt.show()
```
